```python
# @title Default title text
import pandas as pd
i_filepath='/content/MyGoogleDrive/My Drive/Python_Learning/'
df= pd.read_csv(i_filepath +'penguin.csv')
df.info
```

```python
df.drop_duplicates()
```

```python
df.reset_index(drop=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
Species_no = {'Chinstrap penguin':0,'Little penguin':1,'Galapagos penguin':2,'Gentoo penguin':3}
df['Species']=df['Species'].map(Species_no)
#print(df['Species'])
df
```

```python
SpeciesMeanLength = df.groupby(['Species'])['Length_cm'].mean()
SpeciesMeanLength
```

```python
for j in range (4):
  filter = df['Species'] == j
  filtered_df=df[filter]
  x = SpeciesMeanLength[j]
  filtered_df['Length_cm']= filtered_df['Length_cm'].fillna(x)
  df.update(filtered_df)
df
```

```python
SpeciesMeanWeight = df.groupby(['Species'])['Weight_g'].mean()
SpeciesMeanWeight
```

```python
for j in range (4):
  filter = df['Species'] == j
  filtered_df=df[filter]
  x = SpeciesMeanWeight[j]
  filtered_df['Weight_g']= filtered_df['Weight_g'].fillna(x)
  df.update(filtered_df)
```
